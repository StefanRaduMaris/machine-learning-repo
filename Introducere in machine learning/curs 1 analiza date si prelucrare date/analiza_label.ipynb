Vom citi bibliotecile necesare , voi incerca safac un scatter interactiv utilizand plotly

```python
import pandas as pd 
import plotly 
import plotly.express as px
```

Vom citi fisierul

```python
df=pd.read_csv("IMLP4_01-reviews_labeled.csv")
df.head(10)
```

Sa verificam coloanele goale si forma df-ului

```python
print(df.shape)

df1 = df.dropna()

print(df1.isna().sum())

print(df1['sentiment'].value_counts())

df


```

Facem o functie care va parsa textul

```python
def parse_text(df):
    df['sentiment']=df['sentiment'].apply(lambda x : x.lower())
    return df
parse_text(df1)
print(df1['sentiment'].unique())

df1["lungime_review"]=df1['review'].astype(str).apply(lambda x : len(x))
df1['lungime_review']

```

Acum ar trebui sa afisam grafic numerele

```python
fig=px.scatter(
    data_frame=df1,
    x=df1.index,
    y='lungime_review',
    color='sentiment'
)

fig.show()
```

```python
group=df1.groupby('sentiment').agg(
    number_of_votes=('sentiment','count')
)

fig=px.bar(
    data_frame=group,
    y='number_of_votes',
    color='number_of_votes'
)

fig.show()
```
